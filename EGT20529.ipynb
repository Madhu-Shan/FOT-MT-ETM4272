Method 1 : Background Subtraction

````python
# Subtracting the images
img_sub = cv2.subtract(img_b, img_r)

# Thresholding the image, to obtain binary image
ret, img_sub_thresh = cv2.threshold(img_sub, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# Print the used threshold value
print('Threshold Value:', ret)

# Find contours in the image
contours, hierarchy = cv2.findContours(img_sub_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Sort the contours by size in descending order
contours = sorted(contours, key=lambda x: cv2.contourArea(x), reverse=True)
# Print the contour areas
for i in range(len(contours)):
    print('Contour', i, 'Area:', cv2.contourArea(contours[i]))

# Find the area of the whole image
rows, cols = img_sub_thresh.shape
total_area = rows * cols
print('Total Area:', total_area)

# Calculate the percentage of the largest contour area, rounded to 1 decimal point
percentage = round((cv2.contourArea(contours[0]) / total_area) * 100, 1)
print('Percentage:', percentage, '%')

# Display img_r , img_sub, img_sub_thresh side by side using plt
plt.subplot(131), plt.imshow(img_r, cmap='gray'), plt.title('Rand Img')
plt.subplot(132), plt.imshow(img_sub, cmap='gray'), plt.title('Subtracted Img')
plt.subplot(133), plt.imshow(img_sub_thresh, cmap='gray'), plt.title('Thresholded Image')
plt.show()
````

2. Otsu's thresholding method automates the process of finding an optimal global threshold by analyzing the histogram of a bimodal image. It minimizes the within-class variance, effectively separating two distinct peaks in the histogram. The algorithm can handle noise by applying a Gaussian blur before thresholding, improving results. This method is implemented in OpenCV using the cv.threshold() function with the cv.THRESH_OTSU flag, and it is particularly effective for noisy or low-contrast images.

**Initialization**:

The video file is loaded, and the first frame is stored as the reference background.

**Frame Processing:**

For each subsequent frame:
The frame is subtracted from the background to highlight changes.
The result is converted to grayscale to simplify processing.
Thresholding (using Otsu’s method) is applied to create a binary image, isolating regions with significant differences.

**Contour Detection:**

Contours are extracted from the binary image, representing areas of change.
The largest contour area is calculated, and its size is expressed as a percentage of the total frame area.

**Visualization:**

The original frame, subtracted frame, and thresholded frame are displayed side by side for real-time observation.
A plot of the percentage of the largest contour area over time (frame number) is generated for analysis.


````python
import cv2
import matplotlib.pyplot as plt

def process_video(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Unable to open video file {video_path}")
        return

    ret, background = cap.read()
    if not ret:
        print("Error: Unable to read the first frame.")
        cap.release()
        return

    frame_no = 0
    percentages = []

    # Set positions for the windows
    cv2.namedWindow('Original Frame', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Subtracted Frame', cv2.WINDOW_NORMAL)
    cv2.namedWindow('Thresholded Frame', cv2.WINDOW_NORMAL)

    # Adjust positions (x, y) of the windows
    cv2.moveWindow('Original Frame', 0, 100)          # Left
    cv2.moveWindow('Subtracted Frame', 350, 100)     # Center
    cv2.moveWindow('Thresholded Frame', 700, 100)    # Right

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Subtract the current frame from the background
        img_sub = cv2.subtract(background, frame)

        # Convert to grayscale
        img_sub_gray = cv2.cvtColor(img_sub, cv2.COLOR_BGR2GRAY)

        # Thresholding the image to obtain a binary image
        _, img_sub_thresh = cv2.threshold(img_sub_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Find contours
        contours, _ = cv2.findContours(img_sub_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Calculate the largest contour area percentage
        if contours:
            largest_contour_area = max(cv2.contourArea(c) for c in contours)
            frame_area = frame.shape[0] * frame.shape[1]
            percentage = (largest_contour_area / frame_area) * 100
        else:
            percentage = 0

        percentages.append(percentage)

        # Display the video and processed frames
        cv2.imshow('Original Frame', frame)
        cv2.imshow('Subtracted Frame', img_sub)
        cv2.imshow('Thresholded Frame', img_sub_thresh)

        # Break the loop on pressing 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        frame_no += 1

    cap.release()
    cv2.destroyAllWindows()

    # Plot frame number vs percentage
    plt.plot(range(len(percentages)), percentages)
    plt.xlabel('Frame Number')
    plt.ylabel('Percentage of Largest Contour Area')
    plt.title('Frame Number vs Percentage of Largest Contour Area')
    plt.show()

# Call the function with the video path
video_path = 'demo2_images/liquidVideo.mp4'
process_video(video_path)
````
